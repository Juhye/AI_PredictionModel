```
# 데이터프레임에서 데이터 존재하는 월 정보 가져오는 함수
def getMonthCols(df):
  cols = list()
  for i in range(1,13):
    if(i < 10):
      num = '%02d' %i
      col = 'month_%s' % num
    else:
      col = 'month_%d' % i
    if col in df:
      cols.append(col)
      #data2019['months'] = np.where(data2019['months'].isna(), data2019['{}'].format(col), data2019['months'])
  return cols
```

```
# 'months' 칼럼 생성
data2019['months'] = np.nan
```

```
# 월별 날짜를 months 칼럼으로 복사
cols = getMonthCols(data2019)

for col in data2019[cols]:
  data2019['months'] = np.where(data2019['months'].isna(), data2019['{}'.format(col)], data2019['months'])
```

```
# 월별 데이터 One hot encode
# 데이터 있으면 1, 없으면 0
for col in data2019[cols]:
  data2019['{}'.format(col)] = data2019.apply(lambda x: 0 if pd.isnull(x['{}'.format(col)]) else 1, axis=1)
```

```
def process_date(df):
  global df

  data_dummies = pd.get_dummies(df['date'], prefix="date")
  df = pd.concat([df, data_dummies],axis=1)
  df.drop(['date'],axis=1,inplace=True)

  df.rename()
  return df
```
